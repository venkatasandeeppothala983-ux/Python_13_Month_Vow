```python
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([31000, 58000, 92000, 118000, 155000]) # Real-world "noise"
model = LinearRegression()
model.fit(X, y)
prediction = model.predict([[5]])
print(f"The Predicted Salary for 5 years is: {prediction}")
print(model.coef_)
print(model.intercept_)
```

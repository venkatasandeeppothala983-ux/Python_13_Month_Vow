```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([31000, 58000, 92000, 118000, 155000]) # Real-world "noise"
model = LinearRegression()
model.fit(X, y)
prediction = model.predict([[5]])
print(f"The Predicted Salary for 5 years is: {prediction}")
print(model.coef_)
print(model.intercept_)
plt.scatter(X, y, color='purple', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='AI Best Fit Line')
plt.title(f"Salary Predictor (m={model.coef_[0]:.0f}, c={model.intercept_:.0f})")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
